# Welcome to Matplotlib

```python
from matplotlib import pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

# Plotting the data

```python
plt.plot(x,y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title("Sin wave")
plt.show()
```

```python
# cosin wave
plt.plot(x,z)
plt.xlabel("X axis")
plt.ylabel("Z axis")
plt.title("Cosin wave")
plt.show()
```

```python
# construct a parabola
x = np.linspace(-10,10,20)
y = x**2 
plt.plot(x,y)
plt.show()
```
